## Functions

```python
# def name_of_the_function(input1, input2, ...)
    # ... (operations) ...
    # return output1, output2, ...

def plus_one(var):
    var_plus = var + 1
    return var_plus
```

```python
plus_one(2)
```

```python
# functions can have more than one input and output
# return accepts operations
def plus_one_two_inputs(var1, var2):
    return var1 + 1, var2 + 1
```

```python
out1, out2 = plus_one_two_inputs(1, 4)
print(out1, out2)
```
