### Alarm detection

Develop a function that receives data from a sensor as a Python list and returns the string "Alarm!" if any of the values is greater than a given alarm limit, otherwise return the string "Normal"

Tip: double clicks to reveal!
<!---
Declare a variable with 'Normal'. This will be your default message. Iterate over each value. If any of the values is greater than alarm_limit, change the message in the variable to 'Alarm!'. This variable is the output of the function.
-->

During the exercises, challenge yourself. Read the tips only if necessary.





```python
# suggested structure 
# def trigger_alarm(sensor_data, alarm_limite):
```

```python
def trigger_alarm(sensor_data, alarm_limite):
    # sensor starts with status normal
    alarm = 'Normal'

    # iterate over all list values
    for d in sensor_data:
        # if any of the values is greater than alarm_limit
        if d > alarm_limite:
            # replace status to alarm
            alarm = 'Alarm!'

    return alarm
```

```python
# test the function for the following inputs
sensor_trigger = [1.2, 3, 4.1, 2.2]
alarm_limite = 4

trigger_alarm(sensor_trigger, alarm_limite)
```

```python
# test again for the following inputs
sensor_dont_trigger = [1.2, 3, 3.1, 2.2]
trigger_alarm(sensor_dont_trigger, alarm_limite)
```

### Total weekly production
Develop a function that receives a list of the daily production in a week and returns the total production for the week and its daily average

```python
# suggested structure 
# def weekly_production(production_list):
def weekly_production(production_list):
    # start a counter
    total_week = 0

    # for each of the daysf
    for production_day in production_list:
      # add the daily production
      total_week = total_week + production_day # or total_week += production_day (google Assignment Operators)
    
    # evaluates the mean production by day
    mean = total_week / len(production_list)

    return total_week, mean
```

```python
# test the function for the following list
production__list = [230, 254, 214, 208, 235, 120]

total_week, mean = weekly_production(production_list)
print(total_week, mean)
```
