### Filter by date

Build a function that receives a start and end date and filters a DataFrame based on this time range.

```python
def filter_by_date(df, start, end):
    ind = (df.index > start) & (df.index < end)
    return df[ind]
```

### Data compression algorithm (hard)

Code a compression algorithm that reads a Pandas Series and keeps the values only if the difference between the value and the last saved value is greater than a certain threshold.

Tip:
<!---
1. The first value is always saved, so start a Series called "compressed" with only the first value.
2. Iterate over all indexes using the for loop (for index in ser.index).
3. If the difference between the value of this index and the last saved value is greater than the threshold, save (compressed[index] = value).
4. The difference can be positive or negative, so repeat the if statement or use abs(value - compressed[-1]) to ignore the negative sign in the first statement.
-->


```python
def compress(ser, threshold):
    compressed = ser[[0]]

    for index in ser.index:
        value = ser[index]
        if abs(value - compressed[-1]) > threshold:
            compressed[index] = value

    return compressed
```

Run the function to df['Usage'] and threshold = 7. Save the results in a variable named "compressed".

```python
ser = df['Usage']
threshold = 7

compressed = compress(ser, threshold)
compressed.head()
```

Evaluate the compression ratio (uncompressed size / compressed size)

```python
len(df) / len(compressed) 
```

Resample the original dataset to 1 hour sampling time using mean method. Save the result in a variable named "df_resampled".

```python
df_resampled = df.resample("1h").mean()
```

Evaluate the compression ratio for the resampled data.

```python
len(df) / len(df_resampled)
```

The compression ratio for both compression and resampling are very close (close to 4).


The following code show the original, compressed, and resampled data. 

Can you see any difference in the general pattern of the line after the compression? Do you think resampling or compression kept best the pattern of the original data?

```python
import plotly.graph_objects as go

# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=df['Usage'].index, y=df['Usage'].values, mode='lines', name='Raw', line_shape='hv'))
fig.add_trace(go.Scatter(x=compressed.index, y=compressed.values, mode='lines', name='Compressed', line_shape='hv'))
fig.add_trace(go.Scatter(x=df_resampled['Usage'].index, y=df_resampled['Usage'].values, mode='lines', name='Resampled', line_shape='hv'))

fig.show()
```
