# Visualization Libraries - Class

## Matplotlib

Simple graphics

```python
import numpy as np
x = np.arange(0, 10, 1)

y = x ** 2
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y)
plt.show()
```

```python
# some customizations
plt.plot(x, y, color='green', marker='o', linestyle='--')
plt.show()
```

```python
# add labels and title
plt.plot(x, y, color='green', marker='o', linestyle='--')
plt.xlabel('time')
plt.ylabel('value')
plt.title('A exponential curve')
plt.show()
```

```python
# histogram
y_random = np.random.randn(10000)

plt.hist(y_random, bins=50) 
plt.show()
```

```python
# pie plot
z = np.random.rand(3)
print(z)

plt.pie(z)
plt.show()
```

```python
# box plot
z = np.random.randn(100, 3)
plt.boxplot(z)
plt.plot()
```

## Pandas

Quick visualization and initial data exploration

```python
# load steel plant dataset
import pandas as pd
path = 'https://raw.githubusercontent.com/i40a/datasets/main/steel_plant/mods/1/Steel_industry_data.csv' 
df_raw = pd.read_csv(path, index_col='Date', parse_dates=True)
df_raw.head()
```

```python
# select range for better visualization
import datetime as dt

# first 15 days
end = dt.datetime(2018, 1, 15)
df = df_raw[df_raw.index < end]
```

```python
df['Usage'].plot()
plt.show()
```

```python
# histogram
df.hist()
plt.tight_layout() # avoid overlaping
plt.show()
```

```python
# box plot
df.boxplot()
plt.show()
```

```python
# scatter plot
df.plot.scatter('Usage', 'CO2')
plt.show()
```

## Seaborn

Statistical graphics

```python
# histogram
import seaborn as sns
sns.histplot(df['Usage'])
plt.show()
```

```python
# pair plot
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(df, hue='Day') # hue - map plot aspects to different colors.
plt.show()
```

```python
# box plot
sns.boxplot(data=df)
plt.show()
```

```python
sns.boxplot(x=df['Day'], y=df['Usage'])
plt.show()
```

```python
# correlation plot
sns.heatmap(df.corr(), cmap="RdBu_r", annot=True)
plt.show()
```

## Plotly
Interactive graphics

```python
# line chart
import plotly.express as px
fig = px.line(df_raw, x=df_raw.index, y='Usage')
fig.show()
```

```python
# scatter plot
fig = px.scatter(df_raw, x=df_raw.index, y='Usage', color='Day')
fig.show()
```

```python
fig = px.scatter(df, x='Usage', y='LeadRP', color="CO2")
fig.show()
```

```python
fig = px.scatter_matrix(df)
fig.show()
```

```python
fig = px.histogram(df, x="Day", y="Usage")
fig.show()
```
