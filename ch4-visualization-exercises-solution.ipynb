# Visualization Libraries - Exercises

## Matplotlib

Plot the base 10 logarithm from 1 to 10.

```python
# create numpy arrays
import numpy as np

a = np.arange(1, 11)
loga = np.log10(a)
```

```python
# plot array
import matplotlib.pyplot as plt
plt.plot(a, loga)
plt.show()
```

Customize the chart for:
* color: red
* marker: circle
* line style: dash-dot

Also, provide the legends and title of the graph

You can find everything you need on this [link](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
import matplotlib.pyplot as plt
plt.plot(a, loga, marker='o', color='red', linestyle='-.')
plt.xlabel('time')
plt.ylabel('value')
plt.title('A log curve')
plt.show()
```

Generate 10000 uniformly distributed random points and plot the distribution.

```python
random_data = np.random.rand(10000)
plt.hist(random_data, bins=50)
plt.show()
```

## Pandas

Load the Steel Plant Dataset

```python
import pandas as pd
import datetime as dt

path = 'https://raw.githubusercontent.com/i40a/datasets/main/steel_plant/mods/1/Steel_industry_data.csv' 
df = pd.read_csv(path, index_col='Date', parse_dates=True)
df.head()
```

Select only sunday data and plot a boxplot for all variables.

```python
df_sunday = df[df['Day'] == 'Sunday']

df_sunday.boxplot()
plt.show()
```

Remove tuesday data from database and plot Usage against CO2.

```python
df_not_tuesday = df[df['Day'] != 'Tuesday']
df_not_tuesday.plot.scatter('Usage', 'CO2')
plt.show()
```

## Seaborn

For the first 15 days, use the stripplot and the violin plot to plot Day against Usage. (Google it). What do these graphics mean?

```python
# select first 15 days
end = dt.datetime(2018, 1, 15)
df_15_days = df[df.index < end]
df_15_days.head()

import seaborn as sns

# stripplot
sns.stripplot(x=df_15_days['Day'], y=df_15_days['Usage'])
plt.show()
```

```python
# violine plot
sns.violinplot(x=df_15_days['Day'], y=df_15_days['Usage'])
plt.show()
```

## Plotly

Use the "Pie chart with repeated labels" example on [this link](https://plotly.com/python/pie-charts/) to plot a pie chart of usage for each day for the first 15 days.

```python
import plotly.express as px

fig = px.pie(df_15_days, values='Usage', names='Day')
fig.show()
```
